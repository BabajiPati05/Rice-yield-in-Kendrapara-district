```python
import pandas as pd
from src.preprocessing import load_data, remove_iqr_outliers, prepare_features
from src.models import get_models
from src.cross_validation import cross_validate

DATA_2023 = "data/Model_file.csv"
DATA_2024 = "data/Predict_file.csv"

FEATURES = [
    'Mean_Temp_C', 'NDVI_L8', 'NDVI_S2',
    'Precipitation', 'Soil_Moisture'
]
TARGET = 'Yield_2023'

df = load_data(DATA_2023)
df = remove_iqr_outliers(df, TARGET)

X, y, scaler = prepare_features(df, FEATURES, target=TARGET)

models = get_models()
results = []

for name, model in models.items():
    mean_metrics, std_metrics = cross_validate(model, X, y, k=5)
    mean_metrics["Model"] = name
    results.append(mean_metrics)

results_df = pd.DataFrame(results)
results_df.to_csv("results/cv_metrics.csv", index=False)

best_model_name = results_df.sort_values("R2", ascending=False).iloc[0]["Model"]
best_model = models[best_model_name]

best_model.fit(X, y)

df_2024 = load_data(DATA_2024)
X_2024, _ = prepare_features(df_2024, FEATURES, scaler=scaler)
df_2024["Predicted_Yield_2024"] = best_model.predict(X_2024)
df_2024.to_csv("results/predictions_2024.csv", index=False)

print(f"Best model after CV: {best_model_name}")
```
